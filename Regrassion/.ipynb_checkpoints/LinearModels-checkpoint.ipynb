# Fit

```python
def MSE(x1,x2):
    return np.mean((x1-x2)**2)
```

```python
def fit_an_plot(data, columnsToDrop, target_name, dgr, model, plotIt=False):
    """
    Execute:
        train_test_split
        polinomial features expansion
        logarithmic target transformation
        summary plot
    input:
        data (DataFrame) - data that should be fitted and tested
        columnsToDrop (list of strings) - columns that should be ommited
        target_name (string) - name of a target columng
        dgr (int) - degree of polynomial expansion
        model () - model that should be used to fit the data
        plotIt (boolian) - option to make plots
    output:
        accuracy score
        plots if  plotIt = True:
            scatterplot of "test vs prediction"
            residual data
            statistical distirbution of residual data
    """
    X_train, X_test, y_train, y_test = train_test_split(data.drop(columnsToDrop, axis=1), data[target_name], test_size=0.2, random_state=42)
    
    y_train = np.log1p(y_train)
    y_test = np.log1p(y_test)
    
    poly = PolynomialFeatures(degree=dgr)
    X_train = poly.fit_transform(X_train)
    X_test = poly.transform(X_test)
    
    m = model.fit(X_train, y_train)
    prediction = m.predict(X_test)
    
    score = m.score(X_test, y_test)
    
    prediction = np.exp(prediction)-1
    y_test = np.exp(y_test)-1
    
    if plotIt:
        fig = plt.figure(figsize=(15,6))
        ax1 = plt.subplot(1,3,1)
        ax1.scatter(prediction, y_test)
        ax1.set_title(f"Score: {score}")
        ax1.set_ylabel("test")
        ax1.set_xlabel("prediction")

        ax2 = plt.subplot(1,3,2)
        ax2.plot(y_test.to_numpy()-prediction)
        ax2.set_ylabel("residual")

        ax3 = plt.subplot(1,3,3)
        sns.distplot(y_test.to_numpy()-prediction)
        plt.xlabel("residual")
    
    print (MSE(prediction, y_test))
```

```python
%%time
columns_to_drop = "workingday atemp casual registered count".split()
fit_an_plot(df, columns_to_drop, "count", dgr=4, model=LinearRegression(), plotIt=True)
```

```python
%%time 
columns_to_drop = "workingday atemp casual registered count".split()
fit_an_plot(df, columns_to_drop, "count", dgr=4, model=RandomForestRegressor(min_samples_split=3), plotIt=True)
```

# Build Estimator

```python
class My_Estimator(BaseEstimator, TransformerMixin):

    def __init__(self, model, PolyExpansion):
        self.model = model
        self.poly = PolyExpansion
    
    def fit(self, X,y):
        y = np.log1p(y)
        X = self.poly.fit_transform(X)
        self.m = self.model.fit(X, y)

    def predict(self, X):
        prediction = self.m.predict(X)
        return np.exp(prediction)-1

    def score(self, X, y):
        y = np.log1p(y)
        X = self.poly.transform(X)
        return self.m.score(X, y)
```

```python
columnsToDrop = "atemp casual registered count".split()
target_name = "count"
X_train, X_test, y_train, y_test = train_test_split(df.drop(columnsToDrop, axis=1), df[target_name], test_size=0.2, random_state=42)

estimator = My_Estimator(RandomForestRegressor(n_jobs=2), PolynomialFeatures(degree=4))
estimator.fit(X_train, y_train)
estimator.score(X_test, y_test)
```

# HyperparametersOptimization

```python
columns_to_drop = "holiday atemp casual registered count".split()

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns_to_drop, axis=1), 
    df["count"], test_size=0.2, random_state=42
    )

y_train, y_test = np.log1p(y_train), np.log1p(y_test)

colTransform = ColumnTransformer([
    ("Expansion", PolynomialFeatures(), X_train.columns)
    ])

pipe = Pipeline(steps=[
    ("Transform", colTransform),
    ("Model", RandomForestRegressor())
    ])

model_params = {
    'Transform__Expansion__degree': [4,5,6],
    'Model__min_samples_split': [2,3,5]
    }

gridsearch = GridSearchCV(pipe, model_params)
gridsearch.fit(X_train, y_train)
```

# DNN

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["count"], axis=1), df["count"], random_state=42)
```

```python
# to treat differently categorical and contineouse data
# we split them up

categorical = "season holiday workingday".split()
to_drop = "temp hour".split()
contineouse = "humidity windspeed casual casual registered".split()
```

```python
# scale continouse data
standScale = StandardScaler().fit(X_train[contineouse])
X_train_cont = standScale.transform(X_train[contineouse])
X_test_cont = standScale.transform(X_test[contineouse])
```

```python
# 1 hot encode categorical
X_train_cat = X_train[categorical]
X_test_cat = X_test[categorical]

hot1 = OneHotEncoder(sparse=False).fit(X_train_cat)
X_train_cat = hot1.transform(X_train_cat)
X_test_cat = hot1.transform(X_test_cat)
```

```python
# we further build a neaural network
#     that accepts separetly continteouse and categorical data
# Categorical data is first passed through a Dense layer
#     to search for potential patters.
#     And only after translated to contineouase via ReLU activation
#     as the target is contineos value.
# On the countrary contneouse data is being passed onece throuth ReLU
# Eventually both types of inpues are combined to make predications

K.clear_session
input_cat = layers.Input(shape=[8], name = "categorical_Input")
hidden_cat_1 = layers.Dense(30, activation="sigmoid")(input_cat)
hidden_cat_2 = layers.Dense(30, activation="relu")(hidden_cat_1)

input_cont = layers.Input(shape=[5], name = "contintouse_Input")
hidden_cont_1 = layers.Dense(30, activation="relu")(input_cont)

concat = layers.concatenate([hidden_cat_2, hidden_cont_1])
output = layers.Dense("1", name = "output")(concat)

model = keras.Model(inputs = [input_cat, input_cont], outputs = [output])

model.compile(loss="mse", optimizer=keras.optimizers.SGD(learning_rate=1e-3))
```

```python
# We use earlier stop and save parameters corresponding to the best score
checkPoint = keras.callbacks.ModelCheckpoint("best_of_DNN.h5", save_best_only=True)
earlyStop = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
history = model.fit((X_train_cat, X_train_cont), y_train, epochs=1000,
                    validation_split=0.1, callbacks=[checkPoint, earlyStop])
```

```python
plt.plot(history.history["loss"], "-o")
plt.yscale("log")
```

```python
prediction = model.predict((X_test_cat, X_test_cont)).reshape(y_test.shape[0])
```

```python
# the best result is much better that one we could obtain useing sklearn models
MSE(prediction, y_test)
```
